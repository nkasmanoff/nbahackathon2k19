```python
def teamid_lookup(abr,load_all=False):
    """
    Input a team's abbreviation, returns their team id
    
    If load_all is set to true, returns all teams and their associated ids. 
    """

    from nba_py import team
    nba_teams = team.TeamList().info()
    nba_teams = nba_teams[nba_teams['MAX_YEAR'] == '2019']
    if load_all:
        return nba_teams[['TEAM_ID','ABBREVIATION']]
    
    
    teamid = nba_teams.loc[nba_teams['ABBREVIATION'] == abr]['TEAM_ID']
    try:
        return teamid.values[0]
    except:
        return abr + " not found. Try again! "
    
```

```python
all_teams = teamid_lookup('',load_all=True)
```

```python
resting_df = pd.DataFrame([])
for team_ in all_teams.values:#[0:1]:
    teamid = str(team_[0])
    team_name = team_[1]
    team_rest = team.TeamGeneralSplits(team_id=teamid).days_rest()[['GROUP_VALUE','W_PCT']].copy()
    team_rest.index = [team_name for x in team_rest.values]
    resting_df = resting_df.append(team_rest)
```

```python
resting_df.groupby('GROUP_VALUE').mean()
```

```python
resting_df['Days Rest'] = resting_df['GROUP_VALUE'].apply(lambda z: int(z[0]))
```

```python
resting_df.reset_index(inplace=True)
```

```python
plt.figure(figsize=(20,20))

for t,g in resting_df.groupby(['index']):
    plt.plot(g['Days Rest'],g['W_PCT'],'o--',label = t)
plt.title("Win % vs. Days Rest for all teams in NBA",fontsize=26)
plt.xlabel("Days Rest",fontsize=20)
plt.ylabel("Winning %",fontsize=20)
plt.legend()
```

```python
splits_df = pd.DataFrame()

seasons = ['2010-11','2011-12','2012-13','2013-14','2014-15','2015-16','2016-17','2017-18','2018-19']
for season in seasons:
    for team_ in all_teams.values:#[0:1]:
        teamid = str(team_[0])
        team_name = team_[1]
        team_splits = team.TeamGeneralSplits(team_id=teamid,season=season).overall()
        team_splits.index = [team_name + season[-2:] for x in team_splits.values]
        splits_df = splits_df.append(team_splits)
```

```python
splits_df.columns
```

```python
tsnesplits = splits_df[[ 'FGA', 'FG3A','FTA',
       ]].copy()
```

```python
from sklearn import preprocessing

x = tsnesplits.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
tsnesplits_norm = pd.DataFrame(x_scaled)
```

```python
tsnesplits_norm.columns = tsnesplits.columns
tsnesplits_norm.index = tsnesplits.index
```

```python
from sklearn.manifold import TSNE
```

```python
X_tsne = TSNE(n_components=2).fit_transform(tsnesplits_norm)
```

```python
from matplotlib.text import TextPath
plt.figure(figsize=(35,35))
for i,val in enumerate(X_tsne):
    
    label = TextPath((0,0), str(tsnesplits.index[i]),size=24, linewidth=1)
    if tsnesplits.index[i] in ['GSW17','TOR19','GSW18','CLE16','GSW15','SAS14','MIA13','MIA12','DAL11']:
        
        plt.plot(val[0],val[1],'o',color = 'y',  marker=label, markersize=150, linewidth=0,label='champs')
    else:
        plt.plot(val[0],val[1],'o',color = 'k',  marker=label, markersize=100, linewidth=0)
plt.title("")
```
